### Let's obtain these values back and create a script to automatize it
Start by importing the white noise checkerboard data

```python
wn_data = scipy.io.loadmat('/home/idv-eqs8-pza/IDV_code/Variational_GP/hp_optimization_melide/Data_WN/Matlab_output/checkerboard.data_nocomp.checker.exp17_m5.mat')
print(wn_data.keys())
wn_data['Xell'].shape
wn_data['Yell'].shape
wn_data['EllipseCoor'].shape
wn_data['Spatial'].shape
# Good cells. We take them using matlab indices
good_cells = {}
for key, value in wn_data.items():
    if key == '__header__' or key == '__version__' or key == '__globals__':
        continue
    good_cells[key] = np.take(value, idx_matlab, axis=-1)
```
